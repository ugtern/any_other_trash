```python
import matplotlib.pyplot as plt
import math


fail = open('OlimpGrafik.txt', 'r') #указывать при открытии файла формат чтения необязательно он автоматически открывает в формате чтения

arr_znach = []                   # массив значений у 
j=0

for stroka in fail:
    list_stroka = list (stroka)
    fin_str = ''
    for i in range (21,28):
        fin_str+=list_stroka[i]
    arr_znach.append(float (fin_str))
    fin_str = ''
    #print (arr_znach[j])
    j+=1
    
print(list_stroka)

j=0
arr_time = []                     # массив значений x
for j in range (0,len(arr_znach)):
    arr_time.append(float (j))


proizv_znach = []                 # массив значений прозводной f'(x)=y
for j in range (0,len(arr_znach)):
    if j>0: proizv = (arr_znach[j] - arr_znach[j-1])/(arr_time[j] - arr_time[j-1])
    else :  proizv = 0
    proizv_znach.append(float (proizv))




```

```python

plt.plot(proizv_znach)
plt.show()
```

```python
sum_proizv = 0
nakopleniya_proizvodnoi = []
nomer_tochki = []
last_znach_proizv = 0
for j in range (0,len(arr_znach)):
    if proizv_znach[j] : 
        last_znach_proizv = proizv_znach[j]
        if last_znach_proizv*sum_proizv>=0:
            sum_proizv +=  proizv_znach[j]
        elif last_znach_proizv*sum_proizv<=0 and math.fabs(last_znach_proizv)*5<math.fabs(sum_proizv):
            sum_proizv +=  proizv_znach[j]
        else : 
            nakopleniya_proizvodnoi.append(float (sum_proizv))
            nomer_tochki.append(int(j))
            sum_proizv = 0
print (len(nakopleniya_proizvodnoi))        
```

```python
plt.plot(nakopleniya_proizvodnoi)
plt.show()
```

```python
lucky = 0
unlucky = 0
for j in range (0,len(nakopleniya_proizvodnoi)):
    if math.fabs(nakopleniya_proizvodnoi[j]) > 0.0003:
        if (arr_znach[nomer_tochki[j]+60] - arr_znach[nomer_tochki[j]])>0 and (nakopleniya_proizvodnoi[j] > 0):
            lucky +=1
        elif (arr_znach[nomer_tochki[j]+60] - arr_znach[nomer_tochki[j]])<0 and (nakopleniya_proizvodnoi[j] < 0):
            lucky +=1
        else :
            unlucky += 1
print (lucky,unlucky)
```
