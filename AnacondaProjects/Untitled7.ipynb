```python
import numpy as np
import matplotlib.pyplot as plt
import time
import random

shag = 3600
our_test = 180
our_test_j = 0.0002
first = 0
last = shag
indexinger = 0
our_ti = 0.00005
our_kor_len = 20
all_znach = False
sbros = True
i = 1
j = 0.00005
our_file = 'OlimpGrafik4.txt'
dinamika_mass = np.array([])
dinamik_kor_len = np.array([])
koridor = np.array([])
sr_zn_kor = np.array([])
sr_zn_kor_min = np.array([])
sr_zn_kor_max = np.array([])
effec = np.array([])
mass_of_j = np.array([])
mass_of_i = np.array([])
mass_of_prob_len = np.array([])

def first_step(our_file, first, last):
    mass=[]
    f = open(our_file)
    for line in f:
        line = line.replace('\n', '')
        mass.append([line.split(': ')])

    mass = np.array(mass)

    mass_mod = mass[:,0][:,1]
    mnenie_mass = mass[:,0][:,2]

    mass_mod = np.array(mass_mod, dtype=float)
    mnenie_mass = np.array(mnenie_mass, dtype=int)

    if all_znach==True:
        last = len(mnenie_mass)

    mnenie_mass = mnenie_mass[first:last]
    viborka = mass_mod[first:last]

    return mnenie_mass, viborka

def parram(sbros, mnenie_mass, viborka, first, last, effec, mass_of_j, mass_of_i, our_ti, our_kor_len, mass_of_prob_len):
    if sbros == True:
        indexinger = 0
        stavka = 0
        stroka = ''
        koridor = np.array([])
        sr_zn_kor = np.array([])
        sr_zn_kor_min = np.array([])
        sr_zn_kor_max = np.array([])
        prob_kor = np.array([])
        our_index = np.array([])
        uspech_stavki = np.array([])
        neuspech_stavki = np.array([])
        our_uspech_index = np.array([])
        dinamika_mass = np.array([])
        is_stavka = False
        stavka_vverh = False
        stavka_vniz = False
        #viborka = viborka[first:last]
        
        obshak = np.array([dinam_koridor(viborka, our_kor_len, our_ti, 
                  indexinger, stavka, stroka, 
                  koridor, sr_zn_kor, sr_zn_kor_min, sr_zn_kor_max, 
                  prob_kor, our_index, 
                  uspech_stavki, neuspech_stavki, our_uspech_index, 
                  mnenie_mass, dinamika_mass,
                  is_stavka, stavka_vverh, stavka_vniz,
                  effec, mass_of_j, mass_of_i, mass_of_prob_len)])
    else:
        obshak = np.array([dinam_koridor(viborka, our_kor_len, our_ti, 
                  indexinger, stavka, stroka, 
                  koridor, sr_zn_kor, sr_zn_kor_min, sr_zn_kor_max, 
                  prob_kor, our_index, 
                  uspech_stavki, neuspech_stavki, our_uspech_index, 
                  mnenie_mass, dinamika_mass,
                  is_stavka, stavka_vverh, stavka_vniz,
                  effec, mass_of_j, mass_of_i, mass_of_prob_len)])
        
    return obshak

def dinam_koridor(viborka, our_kor_len, our_ti, 
                  indexinger, stavka, stroka, 
                  koridor, sr_zn_kor, sr_zn_kor_min, sr_zn_kor_max, 
                  prob_kor, our_index, 
                  uspech_stavki, neuspech_stavki, our_uspech_index, 
                  mnenie_mass, dinamika_mass,
                  is_stavka, stavka_vverh, stavka_vniz,
                  effec, mass_of_j, mass_of_i, mass_of_prob_len):
    for stroka in viborka:
        
        koridor = np.append(koridor, stroka)
        sr_zn_kor = np.append(sr_zn_kor, np.mean(koridor))

        sr_zn_kor_min = np.append(sr_zn_kor_min, (np.mean(koridor)/our_ti))
        sr_zn_kor_max = np.append(sr_zn_kor_max, (np.mean(koridor)*our_ti))

        if len(koridor)>our_kor_len:
            koridor = np.delete(koridor, 0)

        if len(prob_kor)==0:
            per_pk = stroka
        else:
            per_pk = prob_kor[len(prob_kor)-1]

        if stroka>sr_zn_kor_max[indexinger] and mnenie_mass[indexinger]<40 and stavka ==0:
                prob_kor = np.append(prob_kor, stroka)
                our_index = np.append(our_index, indexinger)
                stavka = 60
                is_stavka = True
                stavka_vverh = True

        elif stroka<sr_zn_kor_min[indexinger] and mnenie_mass[indexinger]>60 and stavka ==0:
                prob_kor = np.append(prob_kor, stroka)
                our_index = np.append(our_index, indexinger)
                stavka = 60
                is_stavka = True
                stavka_vniz = True

        if stavka>0:
            stavka-=1

        if is_stavka == True and stavka == 0:
            if stavka_vverh == True:
                if per_pk>stroka:
                    uspech_stavki = np.append(uspech_stavki, per_pk)
                    our_uspech_index = np.append(our_uspech_index, indexinger-59)
                if per_pk<stroka:
                    neuspech_stavki = np.append(neuspech_stavki, per_pk)
                stavka_vverh = False
            elif stavka_vniz == True:
                if per_pk<stroka:
                    uspech_stavki = np.append(uspech_stavki, per_pk)
                    our_uspech_index = np.append(our_uspech_index, indexinger-59)
                if per_pk>stroka:
                    neuspech_stavki = np.append(neuspech_stavki, per_pk)
                stavka_vniz = False
            is_stavka = False

        indexinger+=1
        
    if len(prob_kor)>1 and (len(uspech_stavki)-len(neuspech_stavki))>0:
        effec = np.append(effec, int(((len(uspech_stavki)-len(neuspech_stavki))/len(prob_kor))*100))
        mass_of_j = np.append(mass_of_j, our_ti)
        mass_of_i = np.append(mass_of_i, our_kor_len)
        mass_of_prob_len = np.append(mass_of_prob_len, len(prob_kor))
        
    return sr_zn_kor_max, sr_zn_kor_min, effec, mass_of_j, mass_of_i, mass_of_prob_len

def obrabotchk_for_otobr(obshak, viborka, first, last):
    
    sr_zn_kor_min = obshak[0][0][1]
    sr_zn_kor_max = obshak[0][0][0]
    #viborka = viborka[first:last]

    otobr(viborka, sr_zn_kor_min, sr_zn_kor_max)

def otobr(viborka, sr_zn_kor_min, sr_zn_kor_max):
    
    print(len(viborka))

    plt.plot(viborka)
    plt.plot(sr_zn_kor_min)
    plt.plot(sr_zn_kor_max)
    fig_size = plt.rcParams["figure.figsize"]
    fig_size[0] = 24
    fig_size[1] = 12
    plt.rcParams["figure.figsize"] = fig_size
    plt.show()

def recursiya_blet(sbros, mnenie_mass, viborka, first, last, effec, mass_of_j, mass_of_i, our_ti, our_kor_len, mass_of_prob_len):
    i = 1
    j = 0.00005
    while j<our_test:
        while i < our_test:
            obshak = np.array([parram(sbros, mnenie_mass, viborka, first, last, effec, mass_of_j, mass_of_i, our_ti, our_kor_len, mass_of_prob_len)])
            i+=1
            our_kor_len = i

            effec = obshak[0][0][2]
            mass_of_j = obshak[0][0][3]
            mass_of_i = obshak[0][0][4]
            mass_of_prob_len = obshak[0][0][5]

        if len(effec)>0 and int(np.max(effec))>90:
            effec_max = np.max(effec)
            print('эфективность '+str(effec_max))
            our_ti = int(mass_of_j[np.where(effec == effec_max)[0][0]])
            our_kor_len = int(mass_of_i[np.where(effec == effec_max)[0][0]])
            prob_o_kor = int(mass_of_prob_len[np.where(effec == effec_max)[0][0]])
            obshak = np.array([parram(sbros, mnenie_mass, viborka, first, last, effec, mass_of_j, mass_of_i, our_ti, our_kor_len, mass_of_prob_len)])
            obrabotchk_for_otobr(obshak, viborka, first, last)
            print(our_ti, our_kor_len, prob_o_kor)
            j = our_test
            i = our_test
        elif j == (our_test-1):
            effec_max = np.max(effec)
            print('эфективность '+str(effec_max))
            our_ti = int(mass_of_j[np.where(effec == effec_max)[0][0]])
            our_kor_len = int(mass_of_i[np.where(effec == effec_max)[0][0]])
            prob_o_kor = int(mass_of_prob_len[np.where(effec == effec_max)[0][0]])
            obshak = np.array([parram(sbros, mnenie_mass, viborka, first, last, effec, mass_of_j, mass_of_i, our_ti, our_kor_len, mass_of_prob_len)])
            obrabotchk_for_otobr(obshak, viborka, first, last)
            print(our_ti, our_kor_len, prob_o_kor)

        j+=0.000001
        vivod_our_ti = our_ti
        our_ti = j
        if len(effec)>0:
            print(mass_of_j[np.where(effec == np.max(effec))[0][0]], int(mass_of_i[np.where(effec == np.max(effec))[0][0]]), np.max(effec), len(effec), int(mass_of_prob_len[np.where(effec == np.max(effec))[0][0]]))

        i = 1
    j = 0.00005
    i=1
    
    print('end recurs tackt')
    
    return vivod_our_ti, our_kor_len, prob_o_kor

obshak = np.array([first_step(our_file, first, last)])

last = shag

mnenie_mass = obshak[0][0]
viborka = obshak[0][1]

obshak = recursiya_blet(sbros, mnenie_mass, viborka, first, last, effec, mass_of_j, mass_of_i, our_ti, our_kor_len, mass_of_prob_len)

print(obshak)

#sr_zn_kor_min = obshak[0][0][1]
#sr_zn_kor_max = obshak[0][0][0]

#first+=shag
#last+=shag
#otobr(viborka, sr_zn_kor_min, sr_zn_kor_max)
```
